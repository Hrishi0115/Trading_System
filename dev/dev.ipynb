```python
import pandas as pd
import yfinance as yf
import uuid
import random
from enum import Enum

def get_signals(symbol, s, e):
    ticker = yf.Ticker(symbol)
    data = ticker.history(start=s, end=e)
    data.index = data.index.tz_localize(None)
    return data

def random_number():
    return random.randint(5,20)

from backtesting_engine.trading_strategies.moving_average_crossover import MovingAverageCrossoverStrategy

mavg_strategy = MovingAverageCrossoverStrategy(short_window=40, long_window=100)

stocks = ["AAPL", "MSFT", "NVDA"]

start = "2020-01-01"
end = "2024-01-01"
signals_dict = {}

for stock in stocks:
    signals_dict[stock] = mavg_strategy.generate_signals(get_signals(stock, start, end))

backtest_timeframe = signals_dict[stocks[0]].index

class Book:
    def __init__(self, starting_cash):
        self.cash = starting_cash
        self.positions = {}
        self.positions_value = 0
        self.positions_value_history = [self.positions_value]
        self.nav = self.cash + self.positions_value
        self.nav_history = [self.nav]
        
        self.trade_history = pd.DataFrame(columns=["trade_id", "symbol", "direction", "quantity", "price", "amount", "date"]).set_index("trade_id")
        self.cash_history = [self.cash]
        
# For both cash_history and position_value_history, these values are calculated end-of-day after all trades for different stocks have occurred.
# Also we include our initial values for cash and position_value -> now these are before trading occurs so not really part of the backtest
# the way we would handle this is either ignore these first values, or just add another date to the beginning of visualisations
# or just have it the first day is the first TRADING day, and since we 
# the initial values are essentially the end of day for the previous day - the day not included in the backtest -> need to decide how to handle this?

# Net Asset Value (NAV): value of the book (cash and positions)
# Assuming NAV = GAV for simplicity

# Should introduce a NoSQL database where we persist this data to, rather than having to include history fields

class Position:
    def __init__(self, symbol, quantity, price):
        self.symbol = symbol
        self.quantity = quantity
        self.average_price = price
        self.value = self.quantity * self.average_price

class Signal(Enum):
    BUY = 1.0
    SELL = -1.0

starting_cash = 100000

book = Book(starting_cash)

# def generate_trade_id():
#     return str(uuid.uuid4())

# trade_id = str(uuid.uuid4())

def trade_log_generator(symbol, direction, quantity, price, date):
    return {
        "trade_id": str(uuid.uuid4()),
        "symbol": symbol,
        "direction": direction,
        "quantity": quantity,
        "price": price,
        "amount": quantity * price,
        "date": date
    }

quantity = 100

for i in range(len(signals_dict[stocks[0]])):
    for symbol, signals in signals_dict.items():
        signal = signals['positions'].iloc[i]
        price = signals['price'].iloc[i]
        date = signals.index[i]
        #quantity = 10.0 # determine this dynamically based on info at index i - add a feature - simple for now
        #quantity = random_number()

        #print(symbol, signal, price, date)

        # ignore risk managment for now - will build high-quality risk management engine in the future

        if signal == Signal.BUY.value:
            # Skipping placing an order logic -> assuming order == trade essentially, at the order stage we would reject if insufficient so we are assuming

            # THINK IT IS IMPORTANT WE ADD AN ORDER TO BUY STAGE -> MAIN TO ENSURE WE DONT MAKE TRADES WHEN WE DONT HAVE THE MONEY

            # we have the cash to buy the share
            # Refactor later
            trade_log = trade_log_generator(symbol, "BUY", quantity, price, date) # we might not need this

            # Handle cash
            book.cash -= trade_log["amount"]

            # Handle positions

            # Position already exists
            if symbol in book.positions.keys():
                # We are already long, this is for completeness
                # update existing position
                position = book.positions[symbol]
                prev_quantity = position.quantity
                prev_price = position.average_price

                position.quantity += quantity # position.quantity = new quantity

                # (old_price * old_shares + new_price * new_shares) / position.quantity # new price
                #print(f"prev_price:{prev_price}, prev_quantity:{prev_quantity}, price:{price}, quantity:{quantity}, position.quantity:{position.quantity}")
                position.average_price = ((prev_price * prev_quantity) + (price * quantity)) / position.quantity

            # New position
            else:
                position = Position(symbol=symbol, quantity=quantity, price=price)

            book.positions[symbol] = position

            #print(symbol, quantity, position.quantity, position.average_price, price)

            # Trade history
            trade_df = pd.DataFrame([trade_log]).set_index("trade_id")
            book.trade_history = pd.concat([book.trade_history, trade_df], axis=0)

            #print("BUY", quantity, position.symbol, position.quantity, position.average_price, price)

        elif signal == Signal.SELL.value:
            # Again assuming no short selling, AND a complete EXIT from the LONG position - in other cells we have been much more disciplined about completeness
            # and less reliant on simplifying assumptions

            trade_log = trade_log_generator(symbol, "SELL", quantity, price, date) # at the moment feels quite redudant - actually used to update
            # the book.trade_history

            quantity_owned = book.positions[symbol].quantity

            if quantity == quantity_owned:
                # Complete exit, update book.positions
                del book.positions[symbol]
                # no longer hold said position
            
            else:
                # partial exit
                position = book.positions[symbol]

                # average_price is unchanged
                # What attributes of the position are affected?
                # quantity

                position.quantity -= quantity
            
            #print("SELL", quantity, position.symbol, position.quantity, position.average_price, price)

            # update cash
            book.cash += trade_log["amount"] # will update so we use getters/setters to handle these operations

            # trade_history

            trade_df = pd.DataFrame([trade_log]).set_index("trade_id")
            book.trade_history = pd.concat([book.trade_history, trade_df], axis=0)

    # Updating book statistics (currently for calculating nav at time t) - cash history and position value
    # BUY/SELL agnostic
    # cash value and history
    book.cash_history.append(book.cash)
        
    # position value and history
    # need to calculate current value of all positions

    total_positions_value = sum(pos.value for pos in book.positions.values()) # find a way to do this part better - like a way to not have to recalculate
    # every timestep which we currently are
    # And currently these aggregate statitics are calculated per time step (in our current case per day - when we start using more granular data will be
    # better - especially in the case where different stocks are traded in the same timestep, this won't be reflected in the best way
    book.positions_value = total_positions_value
    book.positions_value_history.append(total_positions_value)

    book.nav = book.positions_value + book.cash
    book.nav_history.append(book.nav)

    #print(book.nav)
print(book.nav_history)


    # for ticker, pos in book.positions.items():
    #     print(symbol, pos.quantity, pos.average_price, price)

# book.cash

# for ticker, pos in book.positions.items():
#     print(ticker, pos.quantity, pos.average_price)

# What happens when pos.quantity is negative, occurs when allowing non-complete exits and long accumulations - handle edge case when extending framework
# Lots of edge cases we need to model

# book.cash

# for ticker, pos in book.positions.items():
#     print(ticker, pos.quantity, pos.average_price, pos.value)

# print(book.positions_value_history)

# # print(book.cash + 19226.56)

# print(len(book.positions_value_history)) 

# print(len(signals_dict["AAPL"]))

# print(len(book.cash_history))

# For both cash_history and position_value_history, these values are calculated end-of-day after all trades for different stocks have occurred.
```
